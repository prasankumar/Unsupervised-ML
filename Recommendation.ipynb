```python
A = [['Milk', 'Onion', 'Chhole', 'Rajma', 'Ghee', 'Dahi'],
     ['Dill', 'Onion', 'Chhole', 'Rajma', 'Ghee', 'Dahi'],
     ['Milk', 'Apple', 'Rajma', 'Ghee'],
     ['Milk', 'Mutter', 'Corn', 'Rajma', 'Dahi'],
     ['Corn', 'Onion', 'Onion', 'Rajma', 'Ice cream', 'Ghee']]
```

```python
#pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
import pandas as pd
A = pd.DataFrame(te.fit_transform(A),columns=te.columns_)
```

```python
A
```

```python
from mlxtend.frequent_patterns import apriori
Q = apriori(A, min_support=0.5, use_colnames=True)
```

```python
Q
```

```python
P = []
for i in Q.itemsets:
    P.append(len(i))
```

```python
P
```

```python
Q['S'] = P
```

```python
Q[Q.S == 2]
```

```python
def confidence(A,x,y):
    num = A[(A[x] == True) & (A[y] == True)].shape[0]
    den = A[(A[x] == True)].shape[0]
    conf = num/den
    return conf
```

```python
confidence(A,"Ghee","Onion")
```

```python
confidence(A,"Ghee","Rajma")
```

```python
def supp(df,i,j):
    num = A[(A[i] == True) & (A[j] == True)].shape[0]
    den = A.shape[0]
    sup = num/den
    return sup
```

```python
rule = []
support = []
conf = []
for i in A.columns:
    for j in A.columns:
        if(i != j):
            rule.append(i + " " + j)
            support.append(supp(A,i,j))
            conf.append(confidence(A,i,j))
```

```python
T = pd.DataFrame([rule,support,conf]).T
T.columns = ["Rule","Supp","Conf"]
```

```python
T
```

```python
T.shape
```

```python
T[(T.Supp > 0.4) & (T.Conf > 0.4)].sort_values(by=["Rule"])
```

```python
def confidence(Q,p,q):
    num = len(Q[(Q[p]==True)&(Q[q]==True)])
    den = len(Q[Q[p]==True])
    con = round(num/den,2)
    return con
```
